```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Covid Analysis - Phase 1.csv')
```

```python
df.info()
```

```python
unique_countries = df['Country/Region'].nunique()
print(f"Number of unique countries: {unique_countries}")
```

```python
null_values = df.isnull().sum()
print("Null values in the dataset:")
print(null_values)
```

```python
max_confirmed_cases_country = df.loc[df['Confirmed'].idxmax(), 'Country/Region']
print(f"Country with maximum confirmed cases: {max_confirmed_cases_country}")
```

```python
max_deaths_country = df.loc[df['Deaths'].idxmax(), 'Country/Region']
print(f"Country with maximum deaths: {max_deaths_country}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the plots
sns.set(style="whitegrid")
# Visualize the relationship between confirmed cases and deaths
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Confirmed', y='Deaths', data=df)
plt.title('Confirmed Cases vs Deaths')
plt.xlabel('Confirmed Cases')
plt.ylabel('Deaths')
plt.show()

# Top 10 countries with the highest confirmed cases
top_10_countries = df.sort_values('Confirmed', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='Confirmed', y='Country/Region', data=top_10_countries)
plt.title('Top 10 Countries with Highest Confirmed Cases')
plt.xlabel('Confirmed Cases')
plt.ylabel('Country/Region')
plt.show()
```

```python
average_cases = df['Confirmed'].mean()
print(f"Average number of confirmed cases across all countries: {average_cases:.2f}")
```

```python
total_deaths = df['Deaths'].sum()
print(f"Total number of deaths: {total_deaths}")
```

```python
total_confirmed_cases = df['Confirmed'].sum()
print(f"Total number of confirmed cases: {total_confirmed_cases}")
```

```python
df['DeathPercentage'] = (df['Deaths'] / df['Confirmed']) * 100
highest_death_percentage_country = df.loc[df['DeathPercentage'].idxmax(), 'Country/Region']
print(f"Country with the highest death percentage: {highest_death_percentage_country}")
```

```python
df['RecoveryPercentage'] = (df['Recovered'] / df['Confirmed']) * 100
highest_recovery_percentage_country = df.loc[df['RecoveryPercentage'].idxmax(), 'Country/Region']
print(f"Country with the highest recovery percentage: {highest_recovery_percentage_country}")
```

```python


# Visualize the top 20 countries with the highest death percentage
top_20_death_percentage = df.sort_values('DeathPercentage', ascending=False).head(20)
plt.figure(figsize=(12, 6))
sns.barplot(x='DeathPercentage', y='Country/Region', data=top_20_death_percentage)
plt.title('Top 20 Countries with Highest Death Percentage')
plt.xlabel('Death Percentage')
plt.ylabel('Country/Region')
plt.show()

# Visualize the top 20 countries with the highest recovery percentage
top_20_recovery_percentage = df.sort_values('RecoveryPercentage', ascending=False).head(20)
plt.figure(figsize=(12, 6))
sns.barplot(x='RecoveryPercentage', y='Country/Region', data=top_20_recovery_percentage)
plt.title('Top 20 Countries with Highest Recovery Percentage')
plt.xlabel('Recovery Percentage')
plt.ylabel('Country/Region')
plt.show()
```
